```python
def logic_inputs(no_of_inputs):
    """ Helper functie voor het genereren van alle input combinaties bij een logische poort. """
    return list(itertools.product([0, 1], repeat=no_of_inputs))
```

```python
def truth_table(no_of_inputs, network):
    """ Helper functie voor het genereren van een waarheidstabel """
    combinations = logic_inputs(no_of_inputs)
    
    table = ""
    for combination in combinations:
        for bit in combination:
            table += f"{bit}\t"
        table += f"{network(combination)}\n"
            
    return table
```
